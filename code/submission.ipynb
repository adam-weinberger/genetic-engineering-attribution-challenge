```
import pickle
import datatable as dt
import tensorflow as tf
import os
import numpy as np
import pandas as pd
from pathlib import Path
from google.colab import drive
drive.mount('/content/drive')
```

```
os.chdir('/content/drive/My Drive/Colab Notebooks/genetic-engineering-attribution-challenge/code/')
```

```
pd.options.display.max_rows = None
pd.options.display.max_columns = None
train_frac = .8
DATA_DIR = Path.cwd().parent / 'data/train_frac_{}'.format(int(train_frac*100))
```

```
X_test_seq = dt.fread(DATA_DIR / 'X_test_pca.csv', fill=True)
X_test_seq = X_test_seq.to_numpy()
X_test_seq = X_test_seq[:,:8000]
X_test_seq.shape
```

```
X_test_cnn = np.load("{}/X_test_cnn.npy".format(DATA_DIR))
X_test_cnn.shape
```

#run data through models

```
seq_model = tf.keras.models.load_model('{}/models/latest_seq_model'.format(DATA_DIR),
                                               compile=True)
```

```
cnn_model = tf.keras.models.load_model('{}/models/latest_cnn_model'.format(DATA_DIR),
                                               compile=True)
```

```
y_test_probs_seq = seq_model.predict_proba(X_test_seq)
# np.savetxt("{}/y_test_probs_seq.csv".format(DATA_DIR), y_test_probs_seq, delimiter=",")
```

```
y_test_probs_cnn = cnn_model.predict_proba(X_test_cnn)
# np.savetxt("{}/y_test_probs_cnn.csv".format(DATA_DIR), y_test_probs_cnn, delimiter=",")
```

```
del X_test_seq, X_test_cnn
```
